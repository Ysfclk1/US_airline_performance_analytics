```python
# Ensure 'FL_DATE' column is datetime type
import pandas as pd
if not pd.api.types.is_datetime64_any_dtype(df["FL_DATE"]):
    df["FL_DATE"] = pd.to_datetime(df["FL_DATE"])

# 1) NA olmayan saatleri seçme
mask = df["CRS_DEP_TIME"].notna()


# 2) HHMM değerini 4 haneli string

time_str = (
    df.loc[mask, "CRS_DEP_TIME"]
      .astype(int)
      .astype(str)
      .str.zfill(4)
)

# 3) Tarih + saat_string

datetime_str = (
    df.loc[mask, "FL_DATE"].dt.strftime("%Y-%m-%d")
    + " "
    + time_str
)

# 4) Stringi datetime’e dönüştürme
# ===========================
df.loc[mask, "CRS_DEP_DATETIME"] = pd.to_datetime(
    datetime_str,
    format="%Y-%m-%d %H%M"
)

# 5) Saat ve dakika ayrı kolonda
df["CRS_DEP_HOUR"]   = df["CRS_DEP_DATETIME"].dt.hour
df["CRS_DEP_MINUTE"] = df["CRS_DEP_DATETIME"].dt.minute

df[["CRS_DEP_TIME", "CRS_DEP_DATETIME", "CRS_DEP_HOUR", "CRS_DEP_MINUTE"]].head()
```
