```python
# Ensure 'FL_DATE' column is datetime type
import pandas as pd
if not pd.api.types.is_datetime64_any_dtype(df_2015["FL_DATE"]):
    df_2015["FL_DATE"] = pd.to_datetime(df_2015["FL_DATE"])

# 1) NA olmayan saatleri seÃ§me
mask = df_2015["CRS_DEP_TIME"].notna()


# 2) HHMM deÄŸerini 4 haneli string

time_str = (
    df_2015.loc[mask, "CRS_DEP_TIME"]
      .astype(int)
      .astype(str)
      .str.zfill(4)
)

# 3) Tarih + saat_string

datetime_str = (
    df_2015.loc[mask, "FL_DATE"].dt.strftime("%Y-%m-%d")
    + " "
    + time_str
)

# 4) Stringi datetimeâ€™e dÃ¶nÃ¼ÅŸtÃ¼rme
# ===========================
df_2015.loc[mask, "CRS_DEP_DATETIME"] = pd.to_datetime(
    datetime_str,
    format="%Y-%m-%d %H%M"
)

# 5) Saat ve dakika ayrÄ± kolonda
df_2015["CRS_DEP_HOUR"]   = df_2015["CRS_DEP_DATETIME"].dt.hour
df_2015["CRS_DEP_MINUTE"] = df_2015["CRS_DEP_DATETIME"].dt.minute

df_2015[["CRS_DEP_TIME", "CRS_DEP_DATETIME", "CRS_DEP_HOUR", "CRS_DEP_MINUTE"]].head()
```
